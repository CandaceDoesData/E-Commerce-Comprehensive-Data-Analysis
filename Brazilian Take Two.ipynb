```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
import psycopg2
import matplotlib
import jupyter
from pathlib import Path
```

```python
engine = create_engine(
    "postgresql+psycopg2://postgres:tnd.V2SZ@localhost:5432/Brazillian E-Commerce"
)
```

```python
base_path = Path("Resources")

files = {
    #"olist_customers_dataset.csv": "customers",
    #"olist_orders_dataset.csv": "orders",
    #"olist_order_payments_dataset.csv": "order_payments",
    #"olist_geolocation_dataset.csv": "geolocations",
    "olist_order_reviews_dataset.csv": "order_reviews",
    #"olist_products_dataset.csv": "products",
    #"olist_sellers_dataset.csv": "sellers",
    #"olist_order_items_dataset.csv": "order_items",
    #"product_category_name_translation.csv": "category_translations"
}

for file_name, table_name in files.items():
    file_path = base_path / file_name
    
    df = pd.read_csv(file_path)
    
    df.to_sql(
        table_name,
        engine,
        schema="raw",
        if_exists="append",
        index=False
    )
    
    print(f"Loaded {table_name}")
```

```python
views = [
    "executive_kpis",
    "monthly_sales",
    "order_summary",
    "orders_by_state",
    "orders_by_city",
    "top_product_categories",
    "top_products",
    "top_sellers",
    "delivery_performance",
    "customer_satisfaction",
    "order_summary_v2",
    "geo_map"
]

for view in views:
    df = pd.read_sql(f"SELECT * FROM analytics.{view}", engine)
    df.to_csv(f"CSV/{view}.csv", index=False)
    
    print(f"Read {view}")
```
